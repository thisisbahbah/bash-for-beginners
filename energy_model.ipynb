```python
data = pd.read_csv('energy_weather_processed.csv')
target_columns = ['Station 3 Feeder Energy (kWh)', 'Lotto Feeder Energy (kWh)', 'Arena Feeder Energy (kWh)', 'Dormitory Feeder Energy (kWh)', 'Main-Camp Feeder Energy (kWh)', 'Daddy G.O Feeder Energy (kWh)', 'RUN Feeder Energy (kWh)']
data[target_columns] = data[target_columns].fillna(method='ffill')
data['date'] = pd.to_datetime(data['date'])
data['year'] = data['date'].dt.year
data['week_of_year'] = data['date'].dt.isocalendar().week.astype(int)
data['day_of_year'] = data['date'].dt.dayofyear

split_index = 365 # Split after the first 365 days

train_data = data.iloc[:split_index]
test_data = data.iloc[split_index:]

features = ['feelslike', 'temp', 'severerisk', 'month', 'day_of_week', 'hour', 'total_feeders', 'is_event', 'year', 'week_of_year', 'day_of_year']
targets = ['Station 3 Feeder Energy (kWh)', 'Lotto Feeder Energy (kWh)', 'Arena Feeder Energy (kWh)', 'Dormitory Feeder Energy (kWh)', 'Main-Camp Feeder Energy (kWh)', 'Daddy G.O Feeder Energy (kWh)', 'RUN Feeder Energy (kWh)']

X_train = train_data[features]
y_train = train_data[targets]
X_test = test_data[features]
y_test = test_data[targets]

print("Training data shape:", train_data.shape)
print("Testing data shape:", test_data.shape)
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

feeder_columns = y_train.columns
rf_models = {}
gbm_models = {}

for feeder in feeder_columns:
    print(f"Training RandomForestRegressor for {feeder}...")
    rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
    rf_model.fit(X_train, y_train[feeder])
    rf_models[feeder] = rf_model
    print(f"Finished training RandomForestRegressor for {feeder}")

    print(f"Training GradientBoostingRegressor for {feeder}...")
    gbm_model = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=42)
    gbm_model.fit(X_train, y_train[feeder])
    gbm_models[feeder] = gbm_model
    print(f"Finished training GradientBoostingRegressor for {feeder}")

print("All models trained.")
```

```python
rf_predictions = {}
gbm_predictions = {}

for feeder, model in rf_models.items():
    rf_predictions[feeder] = model.predict(X_test)
    print(f"Predictions made for {feeder} using RandomForestRegressor.")

for feeder, model in gbm_models.items():
    gbm_predictions[feeder] = model.predict(X_test)
    print(f"Predictions made for {feeder} using GradientBoostingRegressor.")

print("All predictions made.")
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
import numpy as np

def mean_percentage_error(y_true, y_pred):
    return np.mean((y_true - y_pred) / y_true) * 100

def mean_absolute_scaled_error(y_true, y_pred, y_train):
    n = len(y_true)
    d = np.sum(np.abs(np.diff(y_train))) / (len(y_train) - 1)
    if d == 0:
        return np.nan
    return np.mean(np.abs(y_true - y_pred)) / d

evaluation_metrics = {}

for feeder in y_test.columns:
    evaluation_metrics[feeder] = {}

    # Random Forest Evaluation
    rf_mae = np.mean(np.abs(y_test[feeder] - rf_predictions[feeder]))
    rf_mse = mean_squared_error(y_test[feeder], rf_predictions[feeder])
    rf_mape = mean_absolute_percentage_error(y_test[feeder], rf_predictions[feeder])
    rf_mpe = mean_percentage_error(y_test[feeder], rf_predictions[feeder])
    rf_mase = mean_absolute_scaled_error(y_test[feeder], rf_predictions[feeder], y_train[feeder])

    evaluation_metrics[feeder]['Random Forest'] = {
        'MAE': rf_mae,
        'MSE': rf_mse,
        'MAPE': rf_mape,
        'MPE': rf_mpe,
        'MASE': rf_mase
    }

    # Gradient Boosting Evaluation
    gbm_mae = np.mean(np.abs(y_test[feeder] - gbm_predictions[feeder]))
    gbm_mse = mean_squared_error(y_test[feeder], gbm_predictions[feeder])
    gbm_mape = mean_absolute_percentage_error(y_test[feeder], gbm_predictions[feeder])
    gbm_mpe = mean_percentage_error(y_test[feeder], gbm_predictions[feeder])
    gbm_mase = mean_absolute_scaled_error(y_test[feeder], gbm_predictions[feeder], y_train[feeder])


    evaluation_metrics[feeder]['Gradient Boosting'] = {
        'MAE': gbm_mae,
        'MSE': gbm_mse,
        'MAPE': gbm_mape,
        'MPE': gbm_mpe,
        'MASE': gbm_mase
    }

for feeder, metrics in evaluation_metrics.items():
    print(f"--- {feeder} ---")
    for model, values in metrics.items():
        print(f"  {model}:")
        for metric, value in values.items():
            print(f"    {metric}: {value:.4f}")
```

```python
best_models = {}

for feeder, metrics in evaluation_metrics.items():
    rf_mape = metrics['Random Forest']['MAPE']
    gbm_mape = metrics['Gradient Boosting']['MAPE']

    if rf_mape < gbm_mape:
        best_models[feeder] = {'model': 'Random Forest', 'MAPE': rf_mape}
    else:
        best_models[feeder] = {'model': 'Gradient Boosting', 'MAPE': gbm_mape}

print("Best performing model for each feeder based on MAPE:")
for feeder, details in best_models.items():
    print(f"Feeder: {feeder}, Best Model: {details['model']}, MAPE: {details['MAPE']:.4f}")
```

```python
import pandas as pd

# Create a dictionary to hold the tabulated results
tabulated_results = {}

for feeder, metrics in evaluation_metrics.items():
    for model, values in metrics.items():
        if model not in tabulated_results:
            tabulated_results[model] = {}
        tabulated_results[model][feeder] = values

# Create a DataFrame for each model's metrics
for model, feeder_metrics in tabulated_results.items():
    df_metrics = pd.DataFrame(feeder_metrics).T
    print(f"--- {model} Evaluation Metrics ---")
    display(df_metrics)
    print("\n")
```
